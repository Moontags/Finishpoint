# Osoitehaku – kiitos ja testaus
Tässä notebookissa:

- Toteutetaan uusi osoitehaku-funktio
- Kirjoitetaan yksikkötestit
- Ajetaan testit ja tarkastellaan tuloksia

```python
# 1. Kiitosviestin tulostus
print("Kiitos! Uusi osoitehaku on otettu käyttöön ja testataan seuraavaksi.")
```

```python
# 2. Uuden osoitehaun toteutus (mock)
def osoitehaku(input_str):
    """Mock-funktio: palauttaa osoite-ehdotuksia annetun syötteen perusteella."""
    if not input_str or len(input_str) < 3:
        return []
    # Mockataan Google Places API:n vastaus
    ehdotukset = [
        f"{input_str}katu 1, Helsinki",
        f"{input_str}tie 2, Espoo",
        f"{input_str}polku 3, Vantaa"
    ]
    return ehdotukset
```

```python
# 3. Yksikkötestit osoitehaku-funktiolle
import unittest

class TestOsoitehaku(unittest.TestCase):
    def test_ehdotukset_palautuvat(self):
        tulos = osoitehaku("Hel")
        self.assertTrue(len(tulos) > 0)
        self.assertIn("Helkatu 1, Helsinki", tulos)
    
    def test_lyhyt_syote_palauttaa_tyhjia(self):
        self.assertEqual(osoitehaku("He"), [])
        self.assertEqual(osoitehaku(""), [])
    
    def test_eri_syote_palauttaa_oikeat(self):
        tulos = osoitehaku("Esp")
        self.assertIn("Espkatu 1, Helsinki", tulos)
        self.assertIn("Esptie 2, Espoo", tulos)

# Testien ajaminen
def run_tests():
    suite = unittest.TestLoader().loadTestsFromTestCase(TestOsoitehaku)
    unittest.TextTestRunner(verbosity=2).run(suite)
```

```python
# 4. Testien ajaminen ja tulosten tarkastelu
run_tests()
```
